```python
import pandas as pd

import os
import cv2
```

```python
furniture_df = pd.read_csv("furniture_data.csv")

furniture_df
```

```python
furniture_df['Furniture_Category_Type'] = furniture_df['Furniture_Category'] + '-' + furniture_df['Furniture_Type']

categorical_folders = furniture_df["Furniture_Category_Type"].unique()
print(categorical_folders)
```

```python
furniture_count = furniture_df['Furniture_Category_Type'].value_counts()
furniture_below_1000 = furniture_count[furniture_count < 1000]
```

```python
furniture_below_1000_df = furniture_df[furniture_df['Furniture_Category_Type'].isin(furniture_below_1000.index)].reset_index(drop="index")

furniture_below_1000_df
```

```python
furniture_below_1000_df['Furniture_Category_Type'].value_counts()
```

```python
# Convert the Series to a DataFrame
value_counts_df = furniture_below_1000_df['Furniture_Category_Type'].value_counts().reset_index()

# Rename the columns to appropriate names
value_counts_df.columns = ['Furniture_Category_Type', 'Count']

# Calculate the increase factor for each value
increase_factors = [(5000 // value) if value != 0 else 1 for value in value_counts_df['Count']]

# Add the increase factors as a new column to the DataFrame
value_counts_df['Increase_Factor'] = increase_factors

# Optionally, display the DataFrame to verify the changes
print(value_counts_df)
```

```python
# Merge the two DataFrames based on the 'Furniture_Category_Type' column
furniture_below_1000_df = pd.merge(furniture_below_1000_df, value_counts_df, on='Furniture_Category_Type', how='left')

# Drop the 'Count' column
furniture_below_1000_df.drop(columns=['Count'], inplace=True)

# Optionally, display the merged DataFrame to verify the changes
furniture_below_1000_df
```

```python
import itertools


def sliding_window(image_path, saved_path, window_size=(164,164), desired_windows=6):
    image = cv2.imread(image_path)
    image_height, image_width = image.shape[:2]


    image_filename = os.path.splitext(os.path.basename(image_path))[0]

    # Iterate over possible combinations of window size and step size
    for w, h, S in itertools.product(range(window_size[0], image_width+1), range(window_size[1], image_height+1), range(1, max(image_width, image_height)+1)):
        total_windows = ((image_width - w) // S + 1) * ((image_height - h) // S + 1)

        if abs(total_windows - desired_windows) <= 2:  # Allow flexibility
            # Check if the output directory exists
            if not os.path.exists(saved_path):
                os.makedirs(saved_path)

            # Counter for window index
            window_index = 0

            # Slide the window across the image
            for y in range(0, image_height - h + 1, S):
                for x in range(0, image_width - w + 1, S):
                    window = image[y:y + h, x:x + w]
                    # Do something with the window (e.g., process it, save it, etc.)
                    window_filename = os.path.join(saved_path,f"{image_filename}_{window_index}.jpg")
                    cv2.imwrite(window_filename, window)
                    window_index += 1
            break  # Break after finding the desired number of windows

# Apply sliding window technique for each image in the DataFrame
for index, row in furniture_below_1000_df.iterrows():
    image_path = row['Image_Path']
    image_category = row['Furniture_Category']
    image_interior_type = row['Furniture_Type']
    increase_factors = row['Increase_Factor']
    
    saved_parent_path = f"../CrawledData/{image_category}/{image_interior_type}/"

    sliding_window(
        image_path=image_path, 
        saved_path=saved_parent_path,
        desired_windows= increase_factors
    )
```

```python
# Define the root directory where your categorical furniture folders are located
base_dataset_dir = "../CrawledData"

# Function to get all furniture category names within a directory
def get_furiture_categories(directory):
    folders = []
    for item in os.listdir(directory):
        if os.path.isdir(os.path.join(directory, item)):
            folders.append(item)
    return folders

# Function to get the number of categorical furniture folders and the number of folders within each categorical folder
def get_category_info(root_dir):
    categorical_folders = get_furiture_categories(root_dir)
    num_categorical_folders = len(categorical_folders)
    folder_info = {}
    for cat_folder in categorical_folders:
        cat_folder_path = os.path.join(root_dir, cat_folder)
        folders_in_cat = get_furiture_categories(cat_folder_path)
        num_folders_in_cat = len(folders_in_cat)
        folder_info[cat_folder] = {
            "num_folders": num_folders_in_cat,
            "folders": folders_in_cat
        }
    return num_categorical_folders, folder_info
```

```python
# Get the number of categorical folders and the number of folders within each categorical folder
num_classes, folder_info = get_category_info(base_dataset_dir)
print("Number of categorical furnitures:", num_classes)
print("Info within each categorical furnitures:")
for cat_folder, info in folder_info.items():
    print(f" - {cat_folder}: {info['num_folders']} folders")
    print(f"   {info['folders']}")
```

```python
# Function to get the image paths as a dictionary
def get_image_paths_dict(root_dir):
    image_paths_dict = {}
    for cat_folder, info in folder_info.items():
        cat_image_paths = {}
        for fur_type in info['folders']:
            dir_link = os.path.join(base_dataset_dir, cat_folder, fur_type)
            image_paths = [os.path.join(dir_link, img_path) for img_path in os.listdir(dir_link)]
            cat_image_paths[fur_type] = image_paths
        image_paths_dict[cat_folder] = cat_image_paths
    return image_paths_dict

# Get the image paths as a dictionary
image_paths_dict = get_image_paths_dict(base_dataset_dir)

# Print example paths from the dictionary
for cat_folder, fur_types in image_paths_dict.items():
    for fur_type, paths in fur_types.items():
        print(f"Category: {cat_folder}, Fur type: {fur_type}")
        print("Example paths:")
        for path in paths[:3]:  # print first 3 paths for each fur type
            print(path)
        print("...")
```

```python
# Convert the dictionary of image paths into a DataFrame
df_list = []

for cat_folder, fur_types in image_paths_dict.items():
    for fur_type, paths in fur_types.items():
        for path in paths:
            df_list.append({
                "Furniture_Category": cat_folder,
                "Furniture_Type": fur_type,
                "Image_Path": path
            })

# Create DataFrame from the list of dictionaries
furniture_sliding_df = pd.DataFrame(df_list)

# Print the DataFrame
furniture_sliding_df.head(10)
```
